```python
import re
import nltk
import pandas as pd

data = pd.read_csv("train.csv")
df_data = data.applymap(str)

df_data.insert(4,'testing', '')

#############################################################
################# Expanding words ###########################
#############################################################

for columns in df_data.columns:
    df_data[columns] = df_data[columns].str.lower()
    df_data['testing']=df_data['text'].str.replace("n`t", " not")
    df_data['testing']=df_data['testing'].str.replace("`d", " would")
    df_data['testing']=df_data['testing'].str.replace("`m", " am")
    df_data['testing']=df_data['testing'].str.replace("`ve", " have")


#############################################################
################# Working with Punctuation ##################
#############################################################    
df_data.insert(5,'Exclamation', '')

def removeExtra_Duplicated_asterisk(text):
    return re.sub(r'[*]{2,}', '*', text)

def removeExtra_Duplicated_Exclamation(text):
    return re.sub(r'[!]{2,}', '!', text)

def numberOfExclamation(text):
    total_count = 0;
    exclaimation_mark = text.count('!')
    asterisk = text.count('*')
    total_count = exclaimation_mark + asterisk
    return total_count
    
#Exclamation 語氣
for columns in df_data.columns:
    df_data['testing']=df_data['testing'].str.replace(",", "")
    df_data['testing']=df_data['testing'].str.replace(".", "")
    df_data['testing']=df_data['testing'].apply(removeExtra_Duplicated_asterisk)
    df_data['testing']=df_data['testing'].apply(removeExtra_Duplicated_Exclamation)
    df_data['Exclamation']=df_data['testing'].apply(numberOfExclamation)

    
################################################################
################# Removing stopwrods  ##########################
################################################################

import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english')) 
stop_words.remove('not')

for columns in df_data.columns:
    df_data['testing']=df_data['testing'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop_words)]))

df_data
```
